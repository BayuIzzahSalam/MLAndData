```python
df = pd.read_csv('Latest Covid-19 India Status.csv')
df.head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt

df_join = df[['Total Cases', 'Deaths', 'Death Ratio (%)']].head(10)
df_join = df_join.set_index('Total Cases')

#create line plot
df_join.plot(kind='line', figsize=(20,8))
plt.title('spread of covid 19 in India best 10\n', size = 16)
plt.show()
```

```python
#correlation variables
import seaborn as sns

sns.heatmap(df.corr(), annot=True)
```

```python
#using barplot
plt.figure(figsize=(10,8))
sns.barplot(data=df, x ="Total Cases", y="State/UTs")
```

```python
#total case per state
import plotly.express as px

px.bar(df, x="State/UTs", y="Total Cases", color="Death Ratio (%)", title="total case per state")
```

```python
#highest active
highest_active = df[df['Active'] == max(df['Active'])]
highest_active
```

```python
#lowest active
lowest_active = df[df['Active'] == min(df['Active'])]
lowest_active
```
