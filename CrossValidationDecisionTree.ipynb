```python
import sklearn
from sklearn import datasets

#load iris datasets

iris = datasets.load_iris()
```

```python
#definisikan atribut dan label pada dataset

x = iris.data
y = iris.target
```

```python
from sklearn import tree

#membuat model decisiom tree classifier
#clf adalah classifier

clf = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import cross_val_score

#evaluasi performa model dengan cross_val_score
#cv adalah jumlah fold atau lipatan yang dipakai

scores = cross_val_score(clf, x, y, cv=5)
```

```python
#hasil pengujian setiap fold
scores

#dengan hasil yang tinggi untuk setiap fold dapat dipastikan model decision tree memiliki performa yang baik terhadap dataset iris
```
