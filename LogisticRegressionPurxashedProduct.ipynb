```python
#buang kolom yang tidak penting
df = df1.drop(columns=['User ID'])

#jalankan proses one-hot encoding untuk mengubah label jadi numerik
df = pd.get_dummies(df)
df
```

```python
#pisahkan atribut dan label
predict = ['Age', 'EstimatedSalary', 'Gender_Female', 'Gender_Male']
x = df[predict]
y = df['Purchased']
```

```python
#normalisasi data agar tidak beda jauh nilai antar kolom
from sklearn.preprocessing import StandardScaler
scal = StandardScaler()
scal.fit(x)
scal_data = scal.transform(x)
scal_data = pd.DataFrame(scal_data, columns = x.columns)
scal_data.head()
```

```python
from sklearn.model_selection import train_test_split

#bagi data menjadi train dan tes
x_train, x_test, y_train, y_test = train_test_split(scal_data, y, test_size=0.2, random_state=1)
```

```python
from sklearn import linear_model

#train model dengan fungsi fit
model = linear_model.LogisticRegression()
model.fit(x_train, y_train)
```

```python
#uji akurasi model
model.score(x_test, y_test)
```
