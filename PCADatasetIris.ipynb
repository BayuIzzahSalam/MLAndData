```python
from sklearn import datasets
from sklearn.model_selection import train_test_split

#load data iris
iris = datasets.load_iris()
atribut = iris.data
label = iris.target

#bagi dataset menjadi train dan tes
x_train, x_test, y_train, y_test = train_test_split(atribut, label, test_size=0.2)
```

```python
from sklearn import tree

#menghitung akurasi tanpa PCA dengan decision tree
dec_tree = tree.DecisionTreeClassifier()
model_pertama = dec_tree.fit(x_train, y_train)
model_pertama.score(x_test, y_test)
```

```python
from sklearn.decomposition import PCA

#membuat objek PCA dengan 4 principal components
pca = PCA(n_components = 4)

#mengaplikasikan PCA pada dataset
pca_attributes = pca.fit_transform(x_train)

#melihat variance dari setap atributte
pca.explained_variance_ratio_
```

```python
#PCA dengan 2 principal components 
pca = PCA(n_components=2)
x_train_pca = pca.fit_transform(x_train)
x_test_pca = pca.fit_transform(x_test)
```

```python
#uji akurasi pakai PCA
model2 = dec_tree.fit(x_train_pca, y_train)
model2.score(x_test_pca, y_test)
```
