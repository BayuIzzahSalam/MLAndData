```python
#pisahkan antara atribut dengan label
x = df[df.columns[:13]]
y = df['target']
```

```python
from sklearn.preprocessing import StandardScaler

#standarisasi isi atribut
scaler = StandardScaler()
scaler.fit(x)
x = scaler.transform(x)
```

```python
from sklearn.model_selection import train_test_split

#bagi data train dan data test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier

#buat model 
tree_model = DecisionTreeClassifier()

#training model
tree_model.fit(x_train, y_train)
```

```python
#tingkat akurasi model yaitu 70 %
tree_model.score(x_test, y_test)
```

```python
#uji coba model
tree_model.predict([[52, 1, 1, 120, 240, 1, 0, 125, 0, 1.2, 0, 1, 1]])

#array 0 artinya tidak terkena penyakit jantung
```
