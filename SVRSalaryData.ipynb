```python
#membaca data set dan ubah jadi dataframe
data = pd.read_csv('Salary_Data.csv')
```

```python
#melihat info ada data missing atau tidak
data.info()
```

```python
#menampilkan isi data
data.head()
```

```python
import numpy as np

#memisahkan antara atribut dan label
x = data['YearsExperience']
y = data['Salary']

#mengubah bentuk atribut(karena cuma ada 1 atribut, agar bisa diproses model harus diubah)
x = x[:,np.newaxis]
```

```python
from sklearn.svm import SVR

#membuat model dengan parameter C, gamma dan kernel
model = SVR(C=1000, gamma=0.05, kernel='rbf')

#training model 
model.fit(x, y)
```

```python
import matplotlib.pyplot as plt

#visualisasi model
plt.scatter(x, y)
plt.plot(x, model.predict(x))

#hasil dari pplot memperlihatkan bahwa model tidak mampu menyesuaikan data dengan baik
```
