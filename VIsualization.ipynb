### Menampilkan data perusahaan dengan jumlah pegawai kurang dari 500

### Menampilkan data perusahaan dengan jumlah pegawai lebih dari 100

## Step 4: Visualize the data

Gunakan sel kode berikutnya untuk membuat gambar yang menceritakan kisah di balik kumpulan data Anda. Anda dapat menggunakan jenis bagan apa pun (bar chart, line chart, scatter plot, dll_) yang Anda pilih!

### Melakukan visualisasi jumlah pegawai yang dimiliki oleh perusahaan video games  

````python
plt.figure(figsize=(10, 10))
name = df2['Name']
employee = df2['employeeCount']
barplot = plt.barh(name, employee, color = '#FFA500' )
plt.xlabel('Employee')
plt.ylabel('Name')
plt.show()
````

### Melakukan visualisasi perusahaan video games berdasarkan negara 

````python
df.groupby('country')[['Name']].count().plot(kind='barh', color='r')
````

````python
df.groupby('country')[['Name']].count().plot(kind='pie', subplots = True, legend=False)
````

### Menampilkan data gaji dan pengalaman kerja 

````python
dataframe = pd.read_csv('Salary.csv')
````

````python
dataframe
````

### Melakukan visualisasi dengan line chart untuk jumlah gaji dan tahun pengalaman 

````python
plt.figure(figsize=(15, 15))
plt.plot(dataframe['YearsExperience'], dataframe['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.style.use('seaborn-white')
````

````python
plt.figure(figsize=(15, 15))
plt.plot(dataframe['YearsExperience'], dataframe['Salary'], 'b--')
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.style.use('seaborn-white')
````

### Melakukan visualisasi menggunakan scatter plot untuk jumlah gaji dan tahun pengalaman

````python
plt.scatter(dataframe['YearsExperience'], dataframe['Salary'], color='r')
plt.xlabel("Experience", fontsize = 12)
plt.ylabel("Salary", fontsize = 12)
plt.title("Grafik Scatter")
plt.show()
````
